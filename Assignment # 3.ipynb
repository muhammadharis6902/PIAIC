```python
no_of_spaces = 0
characters = 0

def count_words(filename):
    """Count the approximate number of words in a file."""
    try:
        with open(filename,encoding ='utf8') as f_obj:
            contents = f_obj.read()
    except FileNotFoundError:
        msg = "Sorry, the file " + filename + " does not exist."
        print(msg)
    else:
        # Count approximate number of words in the file.
        words = contents.split()
        num_words = len(words)
        global characters
        characters += sum(len(word) for word in words )
        print( str(num_words) + " words.")
        
def count_lines(filename):
    """Count the approximate number of words in a file."""
    try:
        with open(filename,encoding ='utf8') as f_obj:
            contents = f_obj.read()
    except FileNotFoundError:
        msg = "Sorry, the file " + filename + " does not exist."
        print(msg)
    else:
        # Count approximate number of words in the file.
        lines = contents.split("\n")
        num_lines = len(lines)
        print(str(num_lines) + " lines.")

def count_spaces(filename):
    try:
        with open(filename,encoding ='utf8') as f:
            contents = f.read()
    except FileNotFoundError:
        msg = "Sorry, the file " + filename + " does not exist."
        print(msg)
    else:
        spaces = contents.split(" ")
        num_of_spaces = len(spaces)
        global no_of_spaces
        no_of_spaces = len(spaces)
        print(str(num_of_spaces)+" spaces.")
        
def count_paragraphs(filename):
    try:
        with open(filename,encoding ='utf8') as f:
            contents = f.read()
    except FileNotFoundError:
        msg = "Sorry, the file " + filename + " does not exist."
        print(msg)
    else:
        para = contents.split("   ")
        num_of_paras = len(para)
        print(str(num_of_paras)+" paragraphs.")

def count_tabs(filename):
    try:
        with open(filename,encoding ='utf8') as f:
            contents = f.read()
    except FileNotFoundError:
        msg = "Sorry, the file " + filename + " does not exist."
        print(msg)
    else:
        tabs = contents.split("\t")
        num_of_tabs = len(tabs)
        print(str(num_of_tabs)+" Tabs.")    



filename = input("Enter The Filename: ")
print("\nThe File "+filename+ " Has about :")
count_words(filename)
count_lines(filename)
count_spaces(filename)
count_paragraphs(filename)
print(characters," Characters")
no_of_non_spaces = characters - no_of_spaces
print(no_of_non_spaces,"Non-Space Characters ")
count_tabs(filename)
```

```python
from html.parser import HTMLParser

tags = []
class MyHtmlParser(HTMLParser):
   
    def handle_starttag(self,tag,attrs):
        tags.append(tag)
    def handle_endtag(self,tag):
        tags.append(tag)

filename = input("Enter HTML Filename: ")
with open(filename) as f:
    contents = f.read()
    parser = MyHtmlParser()
    parser.feed(contents)

dct = {}

for i in tags:
    if i in dct:
        dct[i] += 1
    else:
        dct[i] = 1
print ("\nThe File ",filename, " has these tags: ")
for key,value in dct.items():
    print(key,":",value)
```
